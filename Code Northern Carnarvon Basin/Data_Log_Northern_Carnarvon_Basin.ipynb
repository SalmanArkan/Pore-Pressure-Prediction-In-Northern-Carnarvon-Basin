# **Import Library and Dataset**

```python
# Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import Dataset
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_excel('/content/drive/MyDrive/Github/Pore Pressure Prediction In Northern Carnarvon Basin/Dataset/WILCOX-1.xlsx')
df1 = pd.read_excel('/content/drive/MyDrive/Github/Pore Pressure Prediction In Northern Carnarvon Basin/Dataset/WILCOX-2.xlsx')
df2 = pd.read_excel('/content/drive/MyDrive/Github/Pore Pressure Prediction In Northern Carnarvon Basin/Dataset/GOODWYN-6.xlsx')
```

# **Log Northern Carnarvon Basin**

## **WILCOX-1**

```python
# Create A Figure & Set Of Subplots
fig, (ax, ax1, ax2, ax3) = plt.subplots(1, 4, figsize=(12, 8), facecolor='#F0EADD')

# Added Top Border
ax4 = ax.twiny()
ax4.xaxis.set_visible(False)
ax5 = ax1.twiny()
ax5.xaxis.set_visible(False)
ax6 = ax2.twiny()
ax6.xaxis.set_visible(False)
ax7 = ax3.twiny()
ax7.xaxis.set_visible(False)

# Plot Caliper Log
ax.plot(df['CALI.IN'], df['DEPTH.FT'], color='purple')
ax.set_xlabel('Caliper (in)')
ax.set_ylabel('DEPTH (FT)')
ax.tick_params(axis='x', colors="k", width=5)
ax.set_ylim(0, 16000)
ax.invert_yaxis()
ax.set_xlim(8, 32)
ax.set_xticks([8, 12, 16, 20, 24, 28, 32])

# Plot Gamma Ray Log
ax1.plot(df['GR.GAPI'], df['DEPTH.FT'], color='green')
ax1.set_xlabel('GR (GAPI)')
ax1.set_ylabel('DEPTH (FT)')
ax1.tick_params(axis='x', colors="k", width=5)
ax1.set_ylim(0, 16000)
ax1.invert_yaxis()
ax1.set_xlim(0, 200)
ax1.set_xticks([0, 50, 100, 150, 200])

# Plot Density Log
ax2.plot(df['RHOB.G/C3'], df['DEPTH.FT'], color='red')
ax2.set_xlabel('RHOB (G/C3)')
ax2.set_ylabel('DEPTH (FT)')
ax2.tick_params(axis='x', colors="k", width=5)
ax2.set_ylim(0, 16000)
ax2.invert_yaxis()
ax2.set_xlim(1.9, 2.9)
ax2.set_xticks([1.9, 2.1, 2.3, 2.5, 2.7, 2.9])

# Plot Sonic Log
ax3.plot(df['DT.US/F'], df['DEPTH.FT'], color='blue')
ax3.set_xlabel('DT (US/F)')
ax3.set_ylabel('DEPTH (FT)')
ax3.tick_params(axis='x', colors="k", width=5)
ax3.set_ylim(0, 16000)
ax3.invert_yaxis()
ax3.set_xlim(140, 40)
ax3.set_xticks([140, 120, 100, 80, 60, 40])

# Set Grid Lines & Other
for ax in [ax, ax1, ax2, ax3]:
    ax.grid(which='major', color='lightgrey', linestyle='-')
    ax.xaxis.set_ticks_position("top")
    ax.xaxis.set_label_position("top")
    ax.spines["top"].set_position(("axes", 1.02))

# Tighten The Layout & Display The Plot
plt.suptitle('WILCOX-1', fontsize=20)
plt.tight_layout()
plt.show()
```
