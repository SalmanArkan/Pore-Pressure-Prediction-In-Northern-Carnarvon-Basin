# **Import Library and Dataset**

```python
# Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import Dataset
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_excel('/content/drive/MyDrive/Github/Pore Pressure Prediction In Northern Carnarvon Basin/Dataset/WILCOX-1.xlsx')
df1 = pd.read_excel('/content/drive/MyDrive/Github/Pore Pressure Prediction In Northern Carnarvon Basin/Dataset/WILCOX-2.xlsx')
df2 = pd.read_excel('/content/drive/MyDrive/Github/Pore Pressure Prediction In Northern Carnarvon Basin/Dataset/GOODWYN-6.xlsx')
df3 = pd.read_excel('/content/drive/MyDrive/Github/Pore Pressure Prediction In Northern Carnarvon Basin/Dataset/WILCOX-1.xlsx')
df4 = pd.read_excel('/content/drive/MyDrive/Github/Pore Pressure Prediction In Northern Carnarvon Basin/Dataset/WILCOX-2.xlsx')
df5 = pd.read_excel('/content/drive/MyDrive/Github/Pore Pressure Prediction In Northern Carnarvon Basin/Dataset/GOODWYN-6.xlsx')
```

# **Log Gamma Ray**

```python
# Create A List Of Dataframes
df_list = [df, df1, df2]

# Create A List Of Well Names
well_names = ["WILCOX-1", "WILCOX-2", "GOODWYN-6"]

# Create A Figure & Set Of Subplots
fig, axes = plt.subplots(1, 3, figsize = (8, 8), facecolor='#F0EADD')

# Loop Over Each Gamma Ray Log
for i, (ax, df, well_name) in enumerate(zip(axes, df_list, well_names)):
    ax.plot(df['GR.GAPI'], df['DEPTH.FT'], label='Log GR', color='green')
    ax.xaxis.label.set_color("k")
    ax.tick_params(axis='x', colors="k", width=5)
    ax.spines["top"].set_edgecolor("k")
    ax.set_xlabel('GR (GAPI)')

    ax.set_ylabel("DEPTH (FT)")
    ax.set_ylim(0, 12000)
    ax.invert_yaxis()
    ax.set_xlim(0, 200)
    ax.set_xticks([0, 50, 100, 150, 200])
    ax.fill_betweenx(df['DEPTH.FT'], df['GR.GAPI'], 40,
                               where=df['GR.GAPI'] > 40, interpolate=True, color='red', label='Shales')
    ax.fill_betweenx(df['DEPTH.FT'], df['GR.GAPI'], 40,
                               where=df['GR.GAPI'] < 40, interpolate=True, color='yellow', label='Sands')
    ax.axvline(40, color='k', linewidth=1, linestyle='--')
    ax.spines["top"].set_position(("axes", 1.02))
    ax.legend(loc='upper right')
    ax.set_title(label=well_name, fontsize=20, loc="center", color="k")

# Add Gidlines, Ticks, & Labels
for ax in axes:
    ax.grid(which='major', color='lightgrey', linestyle='-')
    ax.xaxis.set_ticks_position("top")
    ax.xaxis.set_label_position("top")
    ax.spines["top"].set_position(("axes", 1.02))

# Tighten The Layout & Display The Plot
plt.tight_layout()
plt.show()
```
